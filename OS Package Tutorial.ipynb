```python
# To create a new folder in the current working directory (Desktop*),
# There is two ways 1st one is
os.mkdir('os_Demo_1')
print(os.listdir())
# Unable to make sub-directories like 'os_Demo_1/os_Demo_1_child1' etc.
# But we can have it by using the second method.
os.makedirs('os_Demo_2/os_Demo_2_child1')
print(os.listdir())
# Able to make sub-directories as well.
```

```python
# To Remove/Delete Directories.
os.rmdir('os_Demo_1')
os.removedirs('os_Demo_2/os_Demo_2_child1')
print(os.listdir())
```

```python
# To rename folder/file.
print(os.listdir())
os.rename('text.txt', 'demo.txt')
print(os.listdir())
```

```python
# To get the details of the file.
print(os.stat('demo.txt'))
```

```python
# To get the modification time of the file.
print(os.stat('demo.txt').st_mtime)
```

```python
# To convert/get the time-stamp to a human read-able format.
from datetime import datetime
mod_time = (os.stat('demo.txt').st_mtime)
print(datetime.fromtimestamp(mod_time))
```
