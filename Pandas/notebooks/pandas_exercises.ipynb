**22. Count total missing values.**

**23. Drop rows with missing values.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    total_missing = df.isna().sum().sum()
    print(f"Total missing values: {total_missing}")

    if total_missing > 0:
        print("Dropping rows with missing values...")
        df = df.dropna()
        print("Rows dropped.")
    else:
        print("No missing values found.")

except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

**24. Fill missing values with mean.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Fill missing values for numeric columns only
    df = df.fillna(df.mean(numeric_only=True))

    print("Numeric missing values filled with mean.")

except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

**25. Fill missing values with median.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print("CSV file read successfully.")

  # Fill missing values for numeric column only.
  df = df.fillna(df.median(numeric_only=True))

  print(f"Numeric values are filled with median.")
except FileNotFoundError:
  print("File not found.")
except Exception as e:
  print(f"An error occured: {e}")
```

**26. Fill missing values with mode.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Fill missing values with mode (for all columns)
    df = df.fillna(df.mode().iloc[0])

    print("Missing values filled with mode.")

except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(f"An error occurred: {e}")
```

**27. Replace negative values with 0.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Replace negative values with 0 only in numeric columns.
    numeric_cols = df.select_dtypes(include=["number"]).columns
    # print(numeric_cols)
    df[numeric_cols] = df[numeric_cols].clip(lower=0)
    print(f"Replaced negative values with 0.")

except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(f"An error occurred: {e}")
```

**29. Remove duplicate rows.**

```python
import pandas as pd

df = pd.read_csv("sample.csv")

# Count duplicate rows
duplicate_count = df.duplicated().sum()

if duplicate_count == 0:
    print("No duplicate rows found.")
else:
    print(f"{duplicate_count} duplicate rows found. Removing them...")
    df.drop_duplicates(inplace=True)
    print("Duplicates removed.")
```

**30. Replace multiple values using a dictionary.**

```python
# Creating a sample dataset

import pandas as pd
import numpy as np

np.random.seed(42)  # for reproducible results

names = [
    "Alice", "Bob", "Charlie", "David", "Emma", "Frank", "Grace", "Helen",
    "Ian", "Jack", "Karen", "Leo", "Mona", "Nora", "Oscar", "Paul", "Queen",
    "Rita", "Sam", "Tina"
]

departments = ["IT", "HR", "Finance", "Sales", "Marketing", None]

data = {
    "employee_id": np.random.randint(100, 200, size=100),   # duplicates likely
    "name": np.random.choice(names, size=100),
    "department": np.random.choice(departments, size=100),  # includes missing
    "salary": np.random.choice([50000, 55000, 60000, 65000, 70000, None], size=100),
    "join_date": pd.to_datetime(
        np.random.choice(pd.date_range("2018-01-01", "2024-01-01"), size=100)
    ),
    "full_time": np.random.choice([True, False, None], size=100)
}

df = pd.DataFrame(data)

print(df.head())
print(df.shape)
```
