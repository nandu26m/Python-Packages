**12. Check memory usage of DataFrame.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print("CSV file read successfully.\n")
  print("Memory usage: ")
  print("-" * 28)
  print(df.memory_usage(deep = True))
except FileNotFoundError:
  print("File not found.")
except Exception as e:
  print(f"An error occured: {e}")
```

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.\n")

    print("Memory usage (per column):")
    print(df.memory_usage(deep=True))

    print("\nTotal memory usage:")
    print(df.memory_usage(deep=True).sum(), "bytes")
except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

**13. Convert a column to int.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print(f"CSV file read successfully.")
  df['track_number'] = df['track_number'].astype(int)
  print("_" * 30)
  print(df.dtypes)
except FileNotFoundError:
  print(f"File not found")
except Exception as e:
  print(f"An error occured: {e}")
```

**14. Convert a column to float.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print(f"CSV file read successfully.")
  df['track_number'] = df['track_number'].astype(float)
  print(df.dtypes)
except FileNotFoundError:
  print(f"File not found")
except Exception as e:
  print(f"An error occured: {e}")
```

**15. Convert a column to datetime.**

```python
import pandas as pd

try:
    # Read the CSV file
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Convert the 'album_release_date' column to datetime
    df['album_release_date'] = pd.to_datetime(df['album_release_date'], errors='coerce')

    # Display the data types of all columns
    print(df.dtypes)

except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
