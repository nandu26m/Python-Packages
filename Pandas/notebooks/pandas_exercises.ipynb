**22. Count total missing values.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print("CSV file read successfully.")
  total_missing = df.isna().sum().sum()
  print(f"Total missing values: {total_missing}")
except FileNotFoundError:
  print("File not found")
except Exception as e:
  print(f"An error occured: {e}")
```

**23. Drop rows with missing values.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    total_missing = df.isna().sum().sum()
    print(f"Total missing values: {total_missing}")

    if total_missing > 0:
        print("Dropping rows with missing values...")
        df = df.dropna()
        print("Rows dropped.")
    else:
        print("No missing values found.")

except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

**24. Fill missing values with mean.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Fill missing values for numeric columns only
    df = df.fillna(df.mean(numeric_only=True))

    print("Numeric missing values filled with mean.")

except FileNotFoundError:
    print("File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

**25. Fill missing values with median.**

```python
import pandas as pd

try:
  df = pd.read_csv("sample.csv")
  print("CSV file read successfully.")

  # Fill missing values for numeric column only.
  df = df.fillna(df.median(numeric_only=True))

  print(f"Numeric values are filled with median.")
except FileNotFoundError:
  print("File not found.")
except Exception as e:
  print(f"An error occured: {e}")
```

**26. Fill missing values with mode.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Fill missing values with mode (for all columns)
    df = df.fillna(df.mode().iloc[0])

    print("Missing values filled with mode.")

except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(f"An error occurred: {e}")
```

**27. Replace negative values with 0.**

```python
import pandas as pd

try:
    df = pd.read_csv("sample.csv")
    print("CSV file read successfully.")

    # Replace negative values with 0 only in numeric columns.
    numeric_cols = df.select_dtypes(include=["number"]).columns
    # print(numeric_cols)
    df[numeric_cols] = df[numeric_cols].clip(lower=0)
    print(f"Replaced negative values with 0.")

except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(f"An error occurred: {e}")
```
