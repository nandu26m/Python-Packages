#A. DataFrame Basics (1â€“20)

```python
import pandas as pd
print(pd.__version__)
```

```python
# Sample data for practice

data = {
    "track_id": ["T001","T002","T003","T004","T005"],
    "track_name": ["Night Sky", "Falling Stars", "Lost Dreams", "Ocean Blue", "Fire Heart"],
    "track_number": [1, 5, 3, 2, 8],
    "track_popularity": [78, 85, 60, 92, 55],
    "track_duration_ms": [210000, 195000, 235000, 180000, 200000],
    "explicit": [False, True, False, False, True],
    "artist_name": ["Astra", "Nova", "Echo", "BlueWave", "Ignis"],
    "artist_popularity": [88, 70, 65, 74, 90],
    "artist_followers": [150000, 95000, 40000, 120000, 300000],
    "artist_genres": [
        ["pop","indie"],
        ["electronic"],
        ["indie","ambient"],
        ["chill","ambient"],
        ["rock","metal"]
    ],
    "album_id": ["AL001","AL002","AL003","AL004","AL005"],
    "album_name": ["Skyline", "Gravity", "Dreamland", "Sea Breeze", "Blaze"],
    "album_release_date": ["2021-06-10", "2020-01-15", "2019-09-20", "2022-02-05", "2023-08-12"],
    "album_total_tracks": [12, 15, 9, 11, 13],
    "album_type": ["album", "single", "album", "album", "single"]
}
```

```python
# Load data to DataFrame
df = pd.DataFrame(data)
```

```python
#1. Show the first 3 rows
df.head(3)
```

```python
#2. Show the last 3 rows
df.tail(3)
```

```python
#3. Print all column names.
df.columns
```

```python
#4. Print DataFrame shape.
df.shape
```

```python
#5. Show DataFrame info.
df.info()
```

```python
#6. Show summary statistics.

# First converting the 'album_release_date' column to 'datetime'
df['album_release_date'] = pd.to_datetime(df['album_release_date'])

print(df.describe(include=['object']))
print(df.describe(include=['number']))
print(df.describe(include=['datetime']))
```

```python
#7. Check number of missing values in each column.
print(df.isnull().sum())
print(df.isna().sum())
```

```python
#8. Convert all column names to uppercase.
df.columns = df.columns.str.upper()
df.columns # The will return an 'Index' object and it has Pandas string-accessor support, so we can perform the above action but its not possible normal lists.
```

```python
#9. Convert all column names to title case.
print(f"Column names before modification:\n {df.columns}\n")
df.columns = df.columns.str.title()
print(f"Column names after modification:\n {df.columns}")
```

```python
#10. Select only numeric columns.
df.select_dtypes(include="number")
```

```python
#11. Select only object columns.
df.select_dtypes(include="object")
```

```python
#12. Change album_release_date to datetime.
if df['Album_Release_Date'].dtypes == "datetime64[ns]":
  print(f"The column 'Album_Release_Date' is already 'datetime64[ns]'.")
else:
  df["Album_Release_Date"] = pd.to_datetime(df["Album_Release_Date"])
  print("The column 'Album_Release_Date' is converted to 'datetime64[ns]'.")
```

```python
#13. Convert track_popularity to float.
df['Track_Popularity'] = df['Track_Popularity'].astype(float)
print(df['Track_Popularity'].dtype)
```

```python
#14. Rename artist_name to singer.
df.rename(columns = {"Artist_Name" : "Singer"}, inplace=True)
df.columns
```

```python
#15. Add a new column track_duration_min (ms â†’ minutes).
df['track_duration_min'] = (df['Track_Duration_Ms'] /1000 / 60).round(2)
print(df.head(5))
```

```python
#16. Drop the column album_id.
print("Before:", df.columns)
df.drop(columns=['Album_Id'], inplace=True)
print("After:", df.columns)
```

```python
#17. Drop first row from the DataFrame.
df.drop(index=df.index[0], inplace=True)
```

```python
df = df.iloc[1:] # Keeps all rows from the second row onward
```

```python
#18. Get DataFrame memory usage.
df.memory_usage(deep=True).sum()

memory = df.memory_usage(deep=True).sum() / (1024 ** 2)
print(f"DataFrame uses {memory:.2f} MB")
```

```python
#19. Check unique values in album_type.
print(df['Album_Type'].unique())
print(df['Album_Type'].nunique())
```

```python
#20. Count number of duplicates.

list_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, list)).any()]
print("Columns containing lists:", list_cols)
```

```python
for col in list_cols:
    df[col] = df[col].apply(lambda x: tuple(x) if isinstance(x, list) else x)
```

```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

#B. Filtering & Querying (21â€“40)

```python
#21. Show tracks with popularity > 80.
df[df['Track_Popularity'] > 80]
```

```python
#22. Show tracks longer than 3 minutes.
df[df['track_duration_min'] > 3]
print(f"Count: {(df['track_duration_min'] > 3).sum()}")
```

```python
#23. Filter rows where explicit == True.
df[df['Explicit'] == True]
```

```python
#24. Get artists with followers > 100,000.
print(df[df['Artist_Followers'] > 100000])
print(f"{'-'*100}\nTotal count: {df[df['Artist_Followers'] > 100000].shape[0]}")
```

```python
#25. Get tracks where artist_genres contains â€œpopâ€.
print(df[df['Artist_Genres'].apply(lambda x: 'rock' in x)].shape[0])
df[df['Artist_Genres'].apply(lambda x: 'rock' in x)]
```

```python
#26. Show albums released after 2021.
df[df['Album_Release_Date'].dt.year > 2021]
```

```python
#27. Get tracks with track_number â‰¤ 3.
df[df["Track_Number"] <= 3]
```

```python
#28. Filter rows where track_name starts with â€œFâ€.
df[df['Track_Name'].str.startswith("F", na=False)]
```

```python
#29. Get rows where track_popularity is between 60 and 90.
df[(df['Track_Popularity'] > 60) & (df['Track_Popularity'] < 90)]
```

```python
df[df["Track_Popularity"].between(60,90, inclusive="neither")]
```

```python
df[df["Track_Popularity"].between(60,90, inclusive="both")]
```

```python
df[df["Track_Popularity"].between(60,90, inclusive="right")]
```

```python
df[df["Track_Popularity"].between(60,90, inclusive="left")]
```

```python
#30. Get tracks where album_type is â€œsingleâ€.
df[df["Album_Type"] == "single"]
```

```python
#31. Filter rows using .query().
df.query("Track_Number == 2")
```

```python
#32. Get rows where artist_popularity < 75.
df[df["Artist_Popularity"] < 75]
```

```python
#33. Get tracks where artist_name is â€œNovaâ€ or â€œEchoâ€.
df[df["Singer"].isin(["Nova", "Echo"])]
```

```python
#34. Show tracks with even track_numbers.
df[df["Track_Number"] % 2 == 0]
```

```python
df.query("Track_Number % 2 == 0")
```

```python
#35. Show rows where album_name contains â€œeâ€.
df[df["Album_Name"].str.contains("e", case=False, na=False)]
```

```python
#36 Filter rows where track_name has more than 8 characters.
df[df["Track_Name"].str.len() > 8]
```
